```python
# Dependencies and Setup
import pandas as pd
import os
import csv
```

```python
# File to Load (Remember to Change These)

#csvpath = os.path.join("Resources", "purchase_data.csv")
#print(csvpath)
purchase_csv = "Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(purchase_csv)
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_df["SN"].unique())
player_count
```

```python
count_df = pd.DataFrame({
    "Player Count": [player_count]
})
count_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# The unique method shows every element of the series that appears only once
unique_item = purchase_df["Item ID"].nunique()
unique_item
```

```python
# The mean method averages the series
average_price = purchase_df["Price"].mean()
average_price
```

```python
number_of_purchases = purchase_df["Purchase ID"].count()
number_of_purchases
```

```python
total_revenue = purchase_df["Price"].sum()
total_revenue
```

```python
# Create a DataFrame of frames using a dictionary of lists
analysis_df = pd.DataFrame({
    "Number of Unique Items": [unique_item],
    "Average Price": [average_price],
    "Number of Purchases": [number_of_purchases], 
    "Total Revenue": [total_revenue]
        
})
analysis_df ["Average Price"] = analysis_df ["Average Price"].map("${:,.2f}".format)
analysis_df ["Total Revenue"] = analysis_df ["Total Revenue"].map("${:,.2f}".format)
analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_total = purchase_df.groupby(["Gender"])
unique_players = gender_total["SN"].nunique()
percentage_of_players = (unique_players / player_count) * 100
```

```python
# Create a DataFrame of frames using a dictionary of lists
demographics_df = pd.DataFrame({
    "Number of Players": unique_players,
    "Percentage of Players": percentage_of_players,
            
})
demographics_df ["Percentage of Players"] = demographics_df ["Percentage of Players"].map("{:,.2f}%".format)
demographics_df
```


## Purchasing Analysis (Gender)

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#create bins
bins = [0,9,14,19,24,29,34,39,60]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# create pd.cut
purchase_df["Age Range"] = pd.cut(purchase_df["Age"], bins, labels = bin_names)
purchase_df
```

```python
age_demographics = purchase_df.groupby("Age Range")
age_unique_totals = age_demographics["SN"].nunique()
age_percent = (age_unique_totals / player_count) * 100
age_percent
```

```python
age_range_df = pd.DataFrame({
    "Total Count": age_unique_totals,
    "Percentage of Players": age_percent
            
})

age_range_df ["Percentage of Players"] = age_range_df ["Percentage of Players"].map("{:,.2f}%".format)
age_range_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_df = purchase_df.groupby(["SN"])
top_spenders_df.first()
```

```python
total_purchase_value = top_spenders_df["Item ID"].count()
total_purchase_value
```

```python
total_price = top_spenders_df["Price"].sum()
average_purchase_price = top_spenders_df["Price"].mean()
```

```python
spenders_analysis_df = pd.DataFrame({
    "Purchase Count": total_purchase_value,
    "Average Purchase Price": average_purchase_price,
    "Total Purchase Value": total_price
})

spenders_analysis_df = spenders_analysis_df.sort_values(["Total Purchase Value"], ascending = False)


spenders_analysis_df ["Average Purchase Price"] = spenders_analysis_df ["Average Purchase Price"].map("${:,.2f}".format)
spenders_analysis_df ["Total Purchase Value"] = spenders_analysis_df ["Total Purchase Value"].map("${:,.2f}".format)
spenders_analysis_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_item_df = purchase_df.groupby(["Item ID", "Item Name"])
popular_purchase_count = popular_item_df["Purchase ID"].count()
popular_item_price = popular_item_df["Price"].mean()
popular_total_price = popular_item_df["Price"].sum()
```

```python
most_popular_df = pd.DataFrame({
    "Purchase Count": popular_purchase_count,
    "Item Price": popular_item_price,
    "Total Purchase Value": popular_total_price
})

most_popular_df = most_popular_df.sort_values(["Total Purchase Value"], ascending = False)

most_popular_df
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
most_profitable_item = purchase_df.groupby(["Item ID", "Item Name"])
most_profitable_count = most_profitable_item["Item ID"].count()
average_purchase_price = most_profitable_item["Price"].mean()
total_purchase_price = most_profitable_item["Price"].sum()
```

```python
most_profitable_df = pd.DataFrame({
    "Purchase Count": most_profitable_count,
    "Item Price": average_purchase_price,
    "Total Purchase Value": total_purchase_price
})

most_profitable_df = most_profitable_df.sort_values(["Total Purchase Value"], ascending = False)

most_profitable_df ["Item Price"] = most_profitable_df ["Item Price"].map("${:,.2f}".format)
most_profitable_df ["Total Purchase Value"] = most_profitable_df ["Total Purchase Value"].map("${:,.2f}".format)

most_profitable_df

```
